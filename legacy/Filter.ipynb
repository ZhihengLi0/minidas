```python
import numpy as np
import matplotlib.pyplot as plt
from ROOT import *
import cats
import supercuts
from CDMSDataCatalog import CDMSDataCatalog
import cdms
import csv
print("CDMS Software Version", cdms.get_global_version()) # print out the CDMS software version
```

```python
rq_path = "/project/rrg-mdiamond/data/CDMS/CUTE/R37/Processed/Releases/CUTE_T3GeCalib_NxM_P4.0.0_V05-06_C0.3.6/Submerged/CUTE_T3GeCalib_NxM_P4.0.0_V05-06_C0.3.6_23231217_135018.root"
df_zip1 = ROOT.RDataFrame("rqDir/zip1", rq_path)
df_evt = ROOT.RDataFrame("rqDir/eventTree", rq_path)

print("== zip1 columns ==")
for col in df_zip1.GetColumnNames():
    print(col)

print("\n== eventTree columns ==")
for col in df_evt.GetColumnNames():
    print(col)
```

```python
event_numbers = df_evt.AsNumpy(["EventNumber"])["EventNumber"]
ptofamps = df_zip1.AsNumpy(["PTOFamps"])["PTOFamps"]
for i in range(10):
    print(f"Index {i}: EventNumber = {event_numbers[i]}, PTOFamps = {ptofamps[i]}")
```

```python
evt_arr = df_evt.AsNumpy(["EventNumber", "DumpNumber", "EventTime", "EventType"])
zip_arr = df_zip1.AsNumpy(["PTOFamps"])
event_numbers = evt_arr["EventNumber"]
dump_numbers = evt_arr["DumpNumber"]
event_times = evt_arr["EventTime"]
event_types = evt_arr["EventType"]
pto_famps = zip_arr["PTOFamps"]

assert len(event_numbers) == len(ptofamps)

filtered = [
    (evt, dump, time, etype, amp)
    for evt, dump, time, etype, amp in zip(event_numbers, dump_numbers, event_times, event_types, pto_famps)
    if etype == 1 and amp > 4e-6 and amp < 9e-6
]

with open("filtered_events.csv", "w", newline='') as f:
    writer = csv.writer(f)
    writer.writerow(["EventNumber", "DumpNumber", "EventTime", "EventType", "PTOFamps"])
    writer.writerows(filtered)
    
print("✅filtered_events.csv")
```
